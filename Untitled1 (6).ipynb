Objective

The purpose of this experiment is to investigate the effectiveness of YOLOv8s with data augmentation for underwater waste detection. The experiment evaluates whether augmentation improves robustness and generalisation in challenging underwater environments.



Code Description

This experiment uses the YOLOv8s architecture with augmentation enabled during training. The model is trained on underwater waste images using transformations such as scaling, flipping, mosaic, and colour variation to improve detection performance.

## Train Final YOLOv8 Model

```python
model = YOLO("yolov8n.pt")

model.train(
    data=data_path,
    epochs=30,
    imgsz=512,
    batch=8,
    device=0,
    augment=True,
    name="final_yolov8n_512_augmented",
    project="/content/drive/MyDrive/yolo_results",
    plots=True
)
```

## Original vs Prediction Comparison

```python
orig_folder = "/content/drive/.shortcut-targets-by-id/1BzlXCMaN335NjAzPSN4OKwengg915o0d/trash_ICRA19/dataset/test/images"

pred_folder = "/content/runs/detect/predict"

images = os.listdir(pred_folder)[:3]

for img_name in images:

    fig, axs = plt.subplots(1,2, figsize=(12,5))

    axs[0].imshow(
        Image.open(os.path.join(orig_folder, img_name))
    )
    axs[0].set_title("Original")

    axs[1].imshow(
        Image.open(os.path.join(pred_folder, img_name))
    )
    axs[1].set_title("Prediction")

    for ax in axs:
        ax.axis("off")

    plt.show()
```

```python
import shutil

output_folder = "/content/member5_good_outputs"

os.makedirs(output_folder, exist_ok=True)

pred_folder = "/content/runs/detect/predict"

pred_images = os.listdir(pred_folder)[:3]

for img in pred_images:

    shutil.copy(
        os.path.join(pred_folder, img),
        os.path.join(output_folder, "prediction_" + img)
    )

train_folder = "/content/runs/detect/member5_final_model"

files_to_copy = [
    "confusion_matrix.png",
    "results.png",
    "labels.jpg"
]

for file in files_to_copy:

    path = os.path.join(train_folder, file)

    if os.path.exists(path):

        shutil.copy(path, output_folder)

print("Saved outputs:", output_folder)
```

Prediction Analysis

The prediction outputs indicate that the augmented YOLOv8s model successfully detected multiple underwater waste categories under varying underwater conditions. The model demonstrated improved robustness in cluttered scenes and maintained stable confidence scores for most visible objects.



 Failure Analysis

Some prediction failures were observed in images with poor lighting, overlapping waste objects, and blurred underwater scenes. In some cases, small objects were missed or detected with lower confidence.

Conclusion

This experiment demonstrated that augmentation improved the generalisation capability of the YOLOv8s model for underwater waste detection. Overall, the model achieved stable performance and improved robustness in challenging underwater environments.
